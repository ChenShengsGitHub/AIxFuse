```python
from rdkit import Chem
from rdkit.Chem import Draw
from rdkit.Chem import rdmolfiles, rdmolops
mol=Chem.MolFromSmiles('CCn1c(CO)nn(-c2ccc3c(=O)n(-c4ccccc4CCC4CCN(C(=O)C(C)C)CC4)cc(C(C)=C)c3c2)c1=O')
smiles=Chem.MolToSmiles(mol,isomericSmiles=False)
mol=Chem.MolFromSmiles(smiles)
new_order = rdmolfiles.CanonicalRankAtoms(mol)
mol = rdmolops.RenumberAtoms(mol, new_order)
for atom in mol.GetAtoms():
    atom.SetProp("atomNote", str(atom.GetIdx()))

Draw.MolToImage(mol,(600,600))
```

```python
import matplotlib.pyplot as plt
from rdkit import Chem
from rdkit.Chem import Draw
from rdkit.Chem import rdmolfiles, rdmolops
from rdkit.Chem.Draw import rdMolDraw2D
# 1. 计算原子权重
weights = [0.1, 0.8, 0.6]  # 假设这是原子的权重，根据实际情况修改
mol = Chem.MolFromSmiles('CCO')  # 输入你的分子结构

# 2. 创建红绿渐变的颜色映射
cmap = plt.cm.get_cmap('YlOrBr')  # 使用RdYlGn颜色映射，即红绿渐变

# 3. 可视化高亮原子
highlight_atoms = {}  # 存储需要高亮显示的原子索引及对应的颜色
for i, atom in enumerate(mol.GetAtoms()):
    if i < len(weights):  # 避免索引超出范围
        color = cmap(1 - weights[i])[:3]  # 根据权重获取对应颜色，并反转权重以获得深浅变化
        highlight_atoms[i] = color
print(highlight_atoms)

Draw.MolToImage(mol, size=(300, 300), highlightAtoms=highlight_atoms)
d = rdMolDraw2D.MolDraw2DCairo(500, 500)
rdMolDraw2D.PrepareAndDrawMolecule(d, mol, highlightAtoms=list(range(3)),highlightAtomColors=highlight_atoms)
d.FinishDrawing()
d.WriteDrawingText('highlighted_molecule.png')
```
